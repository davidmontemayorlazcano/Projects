```python
import sqlite3
conn = sqlite3.connect("RH.db")
```

```python
cur = conn.cursor()
```

```python
#Lectura del archivo CSV
import os
import csv

os.chdir("/Users/Lenovo")
with open('Cientifico-de-Datos-M41-recursos_humanos.csv') as f:
    reader = csv.reader(f)
    data = list(reader)
```

```python
# Creacion de tabla en donde se importara el archivo CSV
cur.execute("""
        CREATE TABLE Detalle (
        satisfaction_level REAL,
        last_evaluation REAL,
        number_project INTEGER,
        average_montly_hours INTEGER,
        time_spend_company INTEGER,
        Work_accident INTEGER,
        left INTEGER,
        promotion_last_5years INTEGER,
        sales TEXT,
        salary TEXT
        );
""")
```

```python
# Insercion de los datos en la tabla recien creada
for renglon in range(1, len(data)):
    cur.execute("""
        INSERT INTO Detalle (satisfaction_level, last_evaluation, number_project, average_montly_hours, time_spend_company, Work_accident, left, promotion_last_5years, sales, salary)
        VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
    """, data[renglon])
```

```python
# Aplicacion de los datos
conn.commit()
```

```python
# Satisfaccion level
cur.execute("SELECT sales, salary, satisfaction_level FROM Detalle ORDER BY satisfaction_level DESC;")
result = cur.fetchall()
result
```

```python
# Ordenamiento conjunto
cur.execute("SELECT salary, number_project, satisfaction_level FROM Detalle ORDER BY number_project ASC, satisfaction_level DESC;")
result = cur.fetchall()
result
```

```python
# Agrupacion Sales
cur.execute("SELECT Sales, AVG(last_evaluation) FROM Detalle GROUP BY sales")
result = cur.fetchall()
result
```

```python
# Agrupacion Sales, Salary
cur.execute("SELECT Sales, salary, AVG(last_evaluation)  FROM Detalle GROUP BY sales, salary")
result = cur.fetchall()
result
```

```python
# Agrupacion Work_accident AVG
cur.execute("SELECT sales, AVG(Work_accident) FROM Detalle GROUP BY sales HAVING AVG(Work_accident)>0.15")
result = cur.fetchall()
result
```

```python
# Agrupacion Work_accident SUM
cur.execute("SELECT sales, SUM(Work_accident) FROM Detalle GROUP BY sales HAVING SUM(Work_accident)>200")
result = cur.fetchall()
result
```
