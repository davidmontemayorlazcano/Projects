```python
import sqlite3
conn = sqlite3.connect("RH.db")
```

```python
cur = conn.cursor()
```

```python
#Lectura del archivo CSV
import os
import csv

os.chdir("/Users/Lenovo")
with open('Cientifico-de-Datos-M41-recursos_humanos.csv') as f:
    reader = csv.reader(f)
    data = list(reader)
```

```python
data
```

```python
len(data)
```

```python
# Creacion de tabla en donde se importara el archivo CSV
cur.execute("""
        CREATE TABLE Detalle (
        satisfaction_level REAL,
        last_evaluation REAL,
        number_project INTEGER,
        average_montly_hours INTEGER,
        time_spend_company INTEGER,
        Work_accident INTEGER,
        left INTEGER,
        promotion_last_5years INTEGER,
        sales TEXT,
        salary TEXT
        );
""")
```

```python
# Insercion de los datos en la tabla recien creada
for renglon in range(1, len(data)):
    cur.execute("""
        INSERT INTO Detalle (satisfaction_level, last_evaluation, number_project, average_montly_hours, time_spend_company, Work_accident, left, promotion_last_5years, sales, salary)
        VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
    """, data[renglon])
```

```python
# Aplicacion de los datos
conn.commit()
```

```python
# Seleccionar columnas especificas
cur.execute("SELECT * FROM Detalle;")
result = cur.fetchall()
result
```

```python
# Satisfaccion Promedio
cur.execute("SELECT AVG(satisfaction_level) FROM Detalle;")
result = cur.fetchall()
result
```

```python
# Satisfaccion Promedio mas alto de quienes deciden dejar la empresa
cur.execute("SELECT MAX(satisfaction_level) FROM Detalle WHERE left = 1;")
result = cur.fetchall()
result
```

```python
# Satisfaccion Promedio mas alto de quienes deciden quedarse en la empresa
cur.execute("SELECT MAX(satisfaction_level) FROM Detalle WHERE left = 0;")
result = cur.fetchall()
result
```

```python
# Horas trabajadas Promedio para salarios LOW
cur.execute("SELECT AVG(average_montly_hours) FROM Detalle WHERE salary = 'low';")
result = cur.fetchall()
result
```

```python
# Horas trabajadas Promedio para salarios MEDIUM
cur.execute("SELECT AVG(average_montly_hours) FROM Detalle WHERE salary = 'medium';")
result = cur.fetchall()
result
```

```python
# Empleados que tuvieron un ascenso en los ultimos 5 años y ademas abandonaron la empresa
cur.execute("SELECT * FROM Detalle WHERE promotion_last_5years = 1 AND left = 1;")
result = cur.fetchall()
result
```

```python
# Empleados que obtuvieron 0.9 o mas en su ultima evaluacion
cur.execute("SELECT * FROM Detalle WHERE last_evaluation >= 0.9;")
result = cur.fetchall()
result
```
