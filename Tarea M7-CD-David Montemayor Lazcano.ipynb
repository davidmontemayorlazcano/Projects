```python
import pandas as pd
import numpy as np
iris = pd.read_csv("Iris.csv")
iris
```

```python
irispetal = iris[['SepalLengthCm','SepalWidthCm']]
irispetal
```

```python
iris
```

```python
condicion1 = iris["SepalLengthCm"] > 5
condicion2 = iris["SepalWidthCm"] >= 3
iris2 = iris[condicion1 & condicion2]
iris2
```

```python
iris2.describe()
```

```python
irispivote = iris2.pivot_table(values = 'PetalWidthCm', index = 'Species', aggfunc = [np.median, np.mean, np.std])
irispivote
```

```python
numeroregistros = iris2['Species'].value_counts()
numeroregistros
```

```python
width = iris2['PetalWidthCm']
petalmean = iris2['PetalWidthCm'].mean()
petalstd = iris2['PetalWidthCm'].std()

iris2['petal.width.st'] = (width - petalmean) / petalstd
iris2
```

```python
iris2.rename(columns = {'Species':'Variedad','petal.width.st':'Ancho Est Petalo'}, inplace = True)
iris2
```

```python
iris2.to_csv('Iris2.csv')
```
